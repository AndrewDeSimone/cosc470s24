```python
from keras.datasets.cifar10 import load_data
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
import numpy as np
import seaborn as sns
```

```python
(trainX, trainy), (testX, testy) = load_data()
# convert from unsigned ints to floats
trainX = trainX.astype('float32')
# scale from [0,255] to [-1,1]
trainX = (trainX - 127.5) / 127.5
```

```python
redtrainX = trainX[:,:,:,0]
greentrainX = trainX[:,:,:,1]
bluetrainX = trainX[:,:,:,2]
redtestX = testX[:,:,:,0]
greentestX = testX[:,:,:,1]
bluetestX = testX[:,:,:,2]
```

```python
fig, ax = plt.subplots(10, 4)
for i in range(0,10):
    ax[i, 0].imshow(trainX[i])
    ax[i, 1].imshow(redtrainX[i])
    ax[i, 2].imshow(greentrainX[i])
    ax[i, 3].imshow(bluetrainX[i])
```

```python
redmodel = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(32,32)),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dense(units=10)
])

redmodel.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True))

losses = redmodel.fit(redtrainX, trainy, epochs=10)
```

```python
bluemodel = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(32,32)),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dense(units=10)
])

bluemodel.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True))

losses = bluemodel.fit(bluetrainX, trainy, epochs=10)
```

```python
greenmodel = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(32,32)),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dense(units=10)
])

greenmodel.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True))

losses = greenmodel.fit(greentrainX, trainy, epochs=10)
```

```python
redPredictions = redmodel.predict(redtestX)
greenPredictions = greenmodel.predict(greentestX)
bluePredictions = bluemodel.predict(bluetestX)
```

```python
redResult = [np.argmax(i) for i in redPredictions]
blueResult = [np.argmax(i) for i in bluePredictions]
greenResult = [np.argmax(i) for i in greenPredictions]
```

```python
#tie goes to red
voterResults = [[0 for j in range(0,10)] for i in range(0,10)]



for i in range(len(testy)):
    if greenResult[i] == blueResult[i]:
        voterResults[blueResult[i]][int(testy[i])] += 1
    else:
        voterResults[redResult[i]][int(testy[i])] += 1

netMap = sns.heatmap(voterResults, annot=True)
netMap.set(xlabel = 'True Values', ylabel = 'Predicted Values', title = 'Voting Network Heat Map')

correctTotal = 0
for i in range(10):
    correctTotal += voterResults[i][i]

print(f'Accuracy: {correctTotal/len(testy)*100}')    
```

```python
cnnmodel = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32,(4,4), input_shape=(32,32,3), activation = 'relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(32,(4,4), input_shape=(32,32,3), activation = 'relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dense(units=10, activation = 'softmax')
])

cnnmodel.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True))
losses = cnnmodel.fit(trainX, trainy, epochs=10)
```

```python
cnnPredictions = cnnmodel.predict(testX)
cnnResult = [np.argmax(i) for i in cnnPredictions]

results = [[0 for j in range(0,10)] for i in range(0,10)]

for i in range(len(testy)):
    results[cnnResult[i]][int(testy[i])] += 1

netMap = sns.heatmap(results, annot=True)
netMap.set(xlabel = 'True Values', ylabel = 'Predicted Values', title = 'CNN Heat Map')

correctTotal = 0
for i in range(10):
    correctTotal += results[i][i]

print(f'Accuracy: {correctTotal/len(testy)*100}')  
```

The cnn performed better overall and had better performance in nearly every category. The voting network had a bias towards one particular class
